```python
import pandas as pd
import matplotlib.pyplot as plt

# === Load your dataset ===
df = pd.read_csv(r"C:\Users\jincy\OneDrive\Desktop\TealProject\Datasets\low-carbon-share-energy (1)\low-carbon-share-energy.csv")

# Ensure correct dtypes
df["Year"] = pd.to_numeric(df["Year"], errors="coerce")
df["Low-carbon energy (% equivalent primary energy)"] = pd.to_numeric(
    df["Low-carbon energy (% equivalent primary energy)"], errors="coerce"
)

# Find latest year in dataset
latest_year = int(df["Year"].max())

# === Plot 1: Trend line for World ===
world_df = df[df["Entity"] == "World"]

plt.figure(figsize=(10, 6))
plt.plot(
    world_df["Year"],
    world_df["Low-carbon energy (% equivalent primary energy)"],
    color="darkblue", linewidth=2
)
plt.title(f"Low-Carbon Energy Share in World (1965{latest_year})", fontsize=14, weight="bold")
plt.xlabel("Year", fontsize=12)
plt.ylabel("Low-carbon energy (% of primary energy)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()

# === Plot 2: Continents trend lines ===
continents = ["Africa", "Asia", "Europe", "North America", "South America", "Oceania"]
cont_df = df[df["Entity"].isin(continents)]

plt.figure(figsize=(12, 7))
for continent in continents:
    subset = cont_df[cont_df["Entity"] == continent]
    plt.plot(
        subset["Year"],
        subset["Low-carbon energy (% equivalent primary energy)"],
        label=continent,
        linewidth=2
    )

plt.title(f"Low-Carbon Energy Share by Continent (1965{latest_year})", fontsize=14, weight="bold")
plt.xlabel("Year", fontsize=12)
plt.ylabel("Low-carbon energy (% of primary energy)", fontsize=12)
plt.legend(title="Continent", loc="upper left", frameon=True)
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```
